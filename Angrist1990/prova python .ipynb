```python
import pandas as pd 
import numpy as np 
```

```python
cwhsa = pd.read_stata('cwhsa.dta')
cwhsc_new = pd.read_stata('cwhsc_new.dta')
cwhsb = pd.read_stata('cwhsb.dta')
```

```python
cwhsa = pd.DataFrame(cwhsa)
cwhsc_new = pd.DataFrame(cwhsc_new)
cwhsb = pd.DataFrame(cwhsb)


cwhsa
```

```python
#per mettere assieme i dataset aggiungo una colonna Type =0 al secondo, riordino per avere la stessa sequenza di colonne e concateno  
cwhsa['type'] = pd.Series([0 for x in range(len(cwhsa['byr']))])
data= cwhsa[['byr', 'race', 'interval','vmn1','vsd1','vnu1','vnms1', 'vfin1','year','ctr1','type','ltax','xltax']]
data=pd.concat([data,cwhsb ], ignore_index=True)
data
```

```python
TAB1 = pd.DataFrame(data)

#rimuovo quelli nati prima del 50 ed estratti prima del 65 

for index, row in TAB1.iterrows(): 
    if row['byr']< 50 or row['year'] <= 65 :
        TAB1.drop(index, inplace=True)
    else : 
        pass
        
TAB1


```

```python
#creo e compilo la colonna sull'elegibilitÃ  
TAB1['elegible'] = pd.Series([0 for x in range(len(TAB1['byr']))])

TAB1['elegible']=np.where(((TAB1['byr']>= 44) & (TAB1['byr']<= 50) & (TAB1['interval']<=39))|((TAB1['byr']==51)  & (TAB1['interval']<=25))|((TAB1['byr']>= 52) & (TAB1['byr']<= 53) & (TAB1['interval']<=59)), 1, 0)
TAB1
```

```python
# aggiungo nella lista le colonne riguardanti le varianze e i dati medi 
TAB1['earn_nz']=TAB1['vmn1']/(1-TAB1['vfin1'])
TAB1['wt_nz']=TAB1['vnu1']*(1-TAB1['vfin1'])
TAB1['WHITE']=np.where(TAB1['race']==1 , 1 , 0 )
TAB1['non WITHE']=np.where(TAB1['race']==2 , 1 , 0 )
TAB1['var']= TAB1['vsd1']*TAB1['vsd1']

TAB1['var_nz']= TAB1['var']*(TAB1['vnu1']/TAB1['wt_nz'])

        
```

```python
TAB1
```

```python
#calcolo i pesi cumulati per gruppo d'appartenenza 
W_g = TAB1.groupby(['race',"byr",'elegible', "year", 'type']).sum().reset_index()

W_g['sumwt'] = W_g['wt_nz']
W_g['wtmult']= 1 / W_g['wt_nz']

W_g = W_g[['race','byr','elegible', 'year', 'type','sumwt','wtmult']]

```

```python
W_g
```

```python
#ristabilisco gli indici e calcolo var_cm e 
W_2 = W_g.merge(TAB1, left_on=['race','byr','elegible', 'year', 'type'], right_on=['race','byr','elegible', 'year', 'type'], how='left')
W_2['var_cm']=W_2['wtmult']*W_2['var_nz']

W_2['type']= np.where(W_2['type']=='TAXAB', 'FICA', 'W')

W_2
```

```python
#calcolo la media dei parametri per gruppo d'appartenenza, e mantengo solo le colonne di mio interesse  
TAB2 = W_2.groupby(['race',"byr", "year", 'type','elegible']).mean().reset_index()

TAB2 = TAB2[['race','byr', 'year', 'type','elegible', 'var_cm','earn_nz']]
```

```python
TAB2
```

```python
#vado a sottrarre fra di loro i dati degli eleggibili e non appartenenti alla stessa classe 
TAB2_b= TAB2.groupby(['race',"byr", "year", 'type']).diff(-1)

TAB2_b
```

```python
#calcolo la radice sulle varianze 
TAB2_b['var_2']= pow(abs(TAB2_b[['var_cm']]), 1/2) 
TAB2_b
```

```python
TAB=TAB2[['race','byr', 'year', 'type']]
```

```python
TAB
```

```python
tab = pd.concat([TAB, TAB2_b], axis=1)
tab=tab[['race','byr', 'year', 'type','earn_nz', 'var_2' ]]
tab=tab.groupby(['race',"byr", "year", 'type']).sum().reset_index()
```
