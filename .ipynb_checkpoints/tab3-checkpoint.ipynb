```python
import pandas as pd 
import numpy as np 

cwhsa = pd.read_stata('cwhsa.dta')
cwhsc_new = pd.read_stata('cwhsc_new.dta')
cwhsb = pd.read_stata('cwhsb.dta')
cpi_angrist1990= pd.read_stata('cpi_angrist1990.dta')
cpi_angrist1990=pd.DataFrame(cpi_angrist1990)
cwhsa = pd.DataFrame(cwhsa)
cwhsc_new = pd.DataFrame(cwhsc_new)
cwhsb = pd.DataFrame(cwhsb)
```

```python
#iweight_old --> iweight
data=cwhsc_new.drop(columns=['iweight'], axis=1)
data['iweight'] = data['iweight_old']
data=data.drop(columns=['iweight_old'], axis=1)
```

```python
data = data.merge(cpi_angrist1990, left_on='year', right_on='year')
```

```python
#trovo le variabili cpi78= media dei cpi per l'anno 78 
i=0
c=0
for index, row in data.iterrows(): 
    if row['year'] ==78 :
        i=i+row['cpi']
        c=c+1
    else : 
        pass

    
cpi78= i/c
cpi78
```

```python
#normalizzo i dati cpi sulla media per l'anno 78 e trovo cpi^2 
data['cpi']=data['cpi'] / cpi78
data['cpi']= round(data['cpi'], 3 )

data['cpi2']=data['cpi']*data['cpi']
data['smplsz']=data['nj']-data['nj0']
```

```python
#tengo solo quelli con year>=81 e byr compresi fra 50 e 53 

for index, row in data.iterrows(): 
    if (row['year'] >= 81) :
        pass        
    else : 
        data.drop(index, inplace=True)

for index, row in data.iterrows(): 
    if  (row['byr'] == 50 or row['byr'] == 51 or row['byr'] == 52):
        pass        
    else : 
        data.drop(index, inplace=True)
```

```python
#trovo varianze e stipendi normalizzati e pongo la variabile di eleggibilit√†

data['rlvar']= (1 / data['iweight']) * data['smplsz']
data['var']= data['rlvar'] * data['cpi2']

data['nomearn']=data['earnings']*data['cpi']

data['elegible'] = pd.Series([0 for x in range(len(data['byr']))])

data['elegible']=np.where(((data['byr']== 50) & (data['interval']<=1))|((data['byr']==51)  & (data['interval']<=25))|((data['byr']== 52) & (data['interval']<=19) | (data['byr']== 53) & (data['interval']<=19)), 1, 0)
```

```python
#applico la comma per gruppi per trovare i pesi cumulati 

W_g = data.groupby(['race',"byr",'elegible', "year", 'type']).sum().reset_index()

W_g['sumwt'] = W_g['smplsz']
W_g['wtmult']= 1 / W_g['sumwt']
W_g = W_g[['race','byr','elegible', 'year', 'type','sumwt','wtmult']]
```

```python
W_2 = W_g.merge(data, left_on=['race','byr','elegible', 'year', 'type'], right_on=['race','byr','elegible', 'year', 'type'], how='left')
W_2['var_cm']=W_2['wtmult'] * W_2['var']
```

```python
#calcolo la media dei parametri per gruppo d'appartenenza, e mantengo solo le colonne di mio interesse  
TAB2 = W_2.groupby(['race',"byr", "year", 'type','elegible']).mean().reset_index()

TAB2 = TAB2[['race','byr', 'year', 'type','elegible', 'var_cm' , 'nomearn', 'cpi', 'earnings']]
```

```python
TAB2
```

```python
#sottrazione per gruppi per trovare le differenze , poi somma per le varianze 
TAB2_b= TAB2.groupby(['race',"byr", "year", 'type']).diff(-1).reset_index()
```

```python
TAB2_var= TAB2.groupby(['race',"byr", "year", 'type']).sum().reset_index()
```

```python
#calcolo la radice sulle varianze 
TAB2_var['var_2']= pow(abs(TAB2_var[['var_cm']]), 1/2) 
```

```python
TAB2_var['var_2']
```

```python
#metto assieme le colonne nelle varie tabelle 

TAB=TAB2[['race','byr', 'year', 'type']]

tab = pd.concat([TAB, TAB2_b], axis=1)
tab=tab[['race','byr', 'year', 'type','nomearn', 'cpi','earnings' ]]
tab=tab.groupby(['race',"byr", "year", 'type']).sum().reset_index()

tab=pd.concat([tab , TAB2_var['var_2']], axis=1)
tab=tab[['race','byr', 'year', 'type','nomearn','var_2']]
```
